### aggregate functions
- count() – Number of non-null observations
- sum() – Sum of values
- mean() – Mean of values
- median() – Arithmetic median of values
- min() – Minimum
- max() – Maximum
- mode() – Mode
- std() – Standard deviation
- var() – Variance

```python
df_nba.groupby("Position")["Salary"].agg([np.mean, np.median])
```

```python
df_nba.groupby("Team")["Age","Weight","Salary"].agg([np.mean, np.median])
```

```python
def height_corr(x):
  feet, inches = x.split("-")
  feet = int(feet)
  inches = int(inches)
  converted_inches = feet*12
  new_height = converted_inches + inches
  return new_height
```

```python
df_nba["Height"] = df_nba["Height"].apply(lambda x : height_corr(x))
```

```python
df_nba["new_Height"] = df_nba["Height"].apply(lambda x : height_corr(x))
```

```python
def height(x):
  feet, inches = x.split("-")
  feet = int(feet)
  inches = int(inches)
  feet = feet * 12
  new_height = feet + inches
  return new_height
```
