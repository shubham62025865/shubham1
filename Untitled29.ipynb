# practics for k_means **alogrithms** 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=1000,centers=4,cluster_std=0.07,random_state=24)
```

```python
x[:5]
```

```python
y[:5]
```

```python
df=pd.DataFrame(data=x,columns=["x1","x2"])
sns.scatterplot(data=df,x="x1",y="x2",hue=y)
plt.show()
```

```python
df.shape
```

```python
for i in range(1,5):
  kmeans=KMeans(n_clusters=4,random_state=0,n_init="auto")
  kmeans.fit(df[["x1","x2"]])
  df["k"+str(i)]=kmeans.labels_

```

```python
df.head()
```

```python
df["k4"] = kmeans.labels_
```

```python
fig, ax = plt.subplots(2,2,figsize = (15,10))
fig.suptitle('Clustering through k means')
# plt.figure(figsize = (15,8))
sns.scatterplot(data = df, x = "x1", y = "x2", hue = "k2", ax = ax[0,0])
ax[0,0].set_title("For k = 2")
sns.scatterplot(data = df, x = "x1", y = "x2", hue = "k3", ax = ax[0,1])
ax[0,1].set_title("For k = 3")
sns.scatterplot(data = df, x = "x1", y = "x2", hue = "k4", ax = ax[1,0])
ax[1,0].set_title("For k = 4")
sns.scatterplot(data = df, x = "x1", y = "x2", hue = y, ax = ax[1,1])
ax[1,1].set_title("Original data clusters")
plt.show()
```

```python
kmeans.inertia_
```

```python
ssd = []
k = []
for i in range(1,10):
  kmeans = KMeans(n_clusters=i, random_state=0, n_init="auto")
  kmeans.fit(df[["x1","x2"]])
  k.append(i)
  ssd.append(kmeans.inertia_)
```

```python
plt.plot(k, ssd)
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(x)
y_kmeans=kmeans.predict(x)
```

```python
plt.scatter(x[:, 0],x[:,1],c=y_kmeans,s=20,cmap="summer")
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c="blue",s=100,alpha=0.9)
plt.show()
```

```python
sse=[]
k_rng=range(1,10)
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(x)
  sse.append(km.inertia_)
```

```python
plt.xlabel("k")
plt.ylabel("sum of squared cluster")
plt.plot(k_rng,sse)
```
