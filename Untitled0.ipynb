## Univariate analysis

- plot histograms of numeric columns
- plot count plot of catgorical columns 

```python
df["work_year"] = df["work_year"].astype("category")
df["remote_ratio"] = df["remote_ratio"].astype("category")
```

```python
num_col = df.dtypes[df.dtypes == "int64"].index
```

```python
# ploting histogram
for col in num_col:
  sns.histplot(df[col], kde = True)
  plt.show()
```

```python
# count plots
cat_cols = df.dtypes[df.dtypes != "int64"]
cat_cols
```

```python
df["experience_level"].value_counts()
```

```python
for col in cat_col:
  num_cat = len(df[col].value_counts())
  if num_cat > 8:
    plt.figure(figsize = (20,8))
    plt.title(f"count plot for {col}")
    sns.countplot(x = df[col], order = df[col].value_counts().index)
    plt.xticks(rotation = 45)
    plt.show()
  else:
    plt.figure(figsize = (10,6))
    plt.title(f"count plot for {col}")
    sns.countplot(x = df[col], order = df[col].value_counts().index)
    plt.xticks(rotation = 0)
    plt.show()
```

## Asking questions

- top 5 job titles for freshers
- average salary for freshers
- top 5 job titles year wise
- top 5 paying jobs for freshers
- number of freshers hired in india

```python
df.columns
```

```python
df["experience_level"].unique()
```

```python
# top 5 job titles for freshers

temp_df = df[df["experience_level"]  == "EN"]
temp_df = temp_df["job_title"].value_counts().head().reset_index(name = "job")
sns.barplot(data = temp_df, x ="index", y = "job" )
plt.xticks(rotation = 45)
plt.show()
```

```python
# for india
temp_df = df[(df["experience_level"]  == "EN") & (df["company_location"] == "IN")]
temp_df = temp_df["job_title"].value_counts().head().reset_index(name = "job")
sns.barplot(data = temp_df, x ="index", y = "job" )
plt.xticks(rotation = 45)
plt.show()
```

```python
temp_df.shape
```

```python
# top 5 highest paying jobs for freshers
temp_df = df[df["experience_level"]  == "EN"]

temp_df = temp_df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending = False).reset_index(name = "count").head()
temp_df

sns.barplot(data = temp_df, x ="job_title", y = "count" )
plt.xticks(rotation = 45)
plt.show()
```
